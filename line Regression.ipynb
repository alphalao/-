```python
class linear:
    def SSErr(self,train_data,train_label):
        train_label=train_label.reshape([train_label.shape[0],1])
        x_=train_data.T.dot(train_data)
        try:
            return np.linalg.inv(x_).dot(train_data.T).dot(train_label)
        except IsADirectoryError:
            return np.linalg.pinv(x_).dot(train_data.T).dot(train_label)
    def LMSalg(self,train_data,train_label,decay=1,rate=0.98,const=1,step=True):
        w=np.random.normal(size=[train_data.shape[1],1])
        train_label=train_label.reshape([train_label.shape[0],1])
        decay_grade=np.zeros([train_data.shape[1],1])
        loss=1000.0
        err_standard=0.0
        while(step):
            loss=train_data.dot(w)-train_label
            loss=np.sum(np.sqrt(loss.T.dot(loss)))/train_data.shape[0]
            print("loss:%g   err_standard:%g"%(loss,err_standard))
            if(abs(loss-err_standard)<0.000000002):
                break
            err_standard=loss
            grade=train_data.T.dot(train_data.dot(w)-train_label)
            decay_grade=decay*decay_grade+(1-decay)*np.multiply(grade,grade)
            w=w-np.multiply(rate/np.sqrt(decay_grade+const),grade)
            if(str(step).isdigit()):
                step=step-1
        return w
```

```python
line=linear()
```

```python
train_data,train_label=data_build(size=[100,100,40,10],label=("random",[1,-1]))
```

```python
print('******************************************homework1*****************************************')
w=line.SSErr(train_data,train_label)
print("w:\n%s"%w)
```

```python
print('******************************************homework2*****************************************')
w=line.LMSalg(train_data,train_label,decay=0.5)
print("w:\n%s"%w)
```
